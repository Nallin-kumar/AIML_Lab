```python
years=[2010,2011,2012,2013,2014,2015]
yield_apples=[0.895,0.91,0.919,0.926,0.929,0.931]
plt.plot(years,yield_apples)
```

```python
plt.plot(years,yield_apples)
plt.xlabel("years")
plt.ylabel("yield (tons per hectare)")
plt.title("Crop yield")
```

```python
import matplotlib.pyplot as plt

year=range(2000,2008)
apples=[0.895,0.91,0.919,0.926,0.929,0.936,0.937, 0.94]
oranges=[0.962,0.941,0.930,0.918,0.908,0.907,0.904,0.901]
plt.plot(year,apples)
plt.plot(year,oranges)
plt.xlabel("years")
plt.ylabel("yield (tons per hectare)")

plt.title("Crop yield")
plt.legend(["Apples","Oranges"])
```

```python
plt.plot(year,apples,marker='o')
plt.plot(year,oranges,marker='x')
plt.xlabel("years")
plt.ylabel("yield (tons per hectare)")

plt.title("Crop yield")
plt.legend(["Apples","Oranges"])
```

```python
plt.plot(year,apples,marker='s',c='b',ls='-',lw=2,ms=8,mew=2,mec='navy')
plt.plot(year,oranges,marker='o',c='r',ls='--',lw=3,ms=10,alpha=0.5)
plt.xlabel("years")
plt.ylabel("yield (tons per hectare)")

plt.title("Crop yield")
plt.legend(["Apples","Oranges"])
```

```python
plt.plot(year,apples,'s-b')
plt.plot(year,oranges,'o--r')
plt.xlabel("years")
plt.ylabel("yield (tons per hectare)")

plt.title("Crop yield")
plt.legend(["Apples","Oranges"])
```

```python
plt.plot(year,oranges,'or')
plt.title('Yield of orange')
```

```python
plt.figure(figsize=(4,4))

plt.plot(year,oranges,'or')
plt.title('Yield of orange')
```

```python
import matplotlib.pyplot as plt

x=[3,1,3,12,2,4,4]
y=[3,2,1,4,5,6,7]

plt.bar(x,y)
plt.title('Bar Chart')
plt.legend(['bar'])
```

```python
plt.scatter(x,y)
plt.title('Scatter Plot')
plt.legend('A')
plt.xlabel('x')
plt.ylabel('y')
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
e=(0.1,0,0,0)
plt.pie(x,explode = e)
plt.title('Pie Chart ')
plt.legend(['Pie'])
plt.show()
```
